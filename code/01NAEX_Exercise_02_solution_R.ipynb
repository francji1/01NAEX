### Exercises 3.07

The tensile strength of Portland cement is being studied. Four different mixing techniques can be used economically. A completely randomized experiment was conducted	and the following data were collected:

| Mixing | Technique Tensile Strength (lb/in2)||||
|--------------------------------------------|||||
| 1      |  3129  |  3000  |  2865  |  2890  |
| 2      |  3200  |  3300  |  2975  |  3150  |
| 3      |  2800  |  2900  |  2985  |  3050  |
| 4      |  2600  |  2700  |  2600  |  2765  |

* Construct a graphical display to compare the mean tensile strengths for the
four mixing techniques. What are your conclusions?
* Test the hypothesis that mixing techniques affect the
strength of the cement. Use  $\alpha = 0.05$.
* Use the Fisher LSD method with  $\alpha = 0.05$ to make
comparisons between pairs of means.
*  Construct a normal probability plot of the residuals.
What conclusion would you draw about the validity of
the normality assumption?
*  Plot the residuals versus the predicted tensile strength. Comment on the plot.
* Prepare a scatter plot of the results to aid the interpretation of the results of this experiment.

```r
Ex03_7 <- read.table("https://raw.githubusercontent.com/francji1/01NAEX/main/data/Ex03_7.csv",header=TRUE,sep=";")
head(Ex03_7)
str(Ex03_7)

library(ggplot2)
library(tidyverse)
install.packages('car')
library(car)
data <- Ex03_7 %>% transmute(stren = Tensile_Strength,
                          tech = as.factor(Technique))
ggplot(data=data, aes(x=tech, y=stren)) +
  geom_boxplot() +
  stat_summary(fun.y="mean")

```

The conclusions are that all mixing techniques have different mean tensile strength except techniques 1 and 3. Further, all mixing techniques have similar variance of tensile strength.

```r
data
```

```r
anova_result <- aov(stren ~ tech, data=data)
summary(anova_result)
```

The p-value is lower than $α=0.05$, so we reject the null hypothesis, which means that the mixing technique has an effect on the strength.

```r
lsd_result <- LSD.test(anova_result, "tech", alpha=0.05, group=FALSE)
print(lsd_result)
```

  From pair-wise comparison of means using Fisher LSD test, we see that there is a significant difference at $α=0.05$ level between all methods except mixing techniques 1 and 3.

```r
qqnorm(anova_result$residuals)
qqline(anova_result$residuals)
```

According to the QQ plot, the residuals appear to have an approximately normal distribution.

```r
predicted_values <- fitted(anova_result)

plot(predicted_values, anova_result$residuals, xlab="Predicted Tensile Strength", ylab="Residuals")
abline(h=0, col="red")
```

Some pattern can be seen in the graph, indicating problems with the assumptions or that there are opportunities to improve the model.

```r
ggplot(data, aes(x=tech, y=stren)) +
  geom_point(aes(color=tech), position=position_jitter(width=0.2)) +
  labs(title="Tensile Strength by Mixing Technique", x="Mixing Technique", y="Tensile Strength (lb/in2)")
```

The graph supports the results of the Fisher LSD test and that the different mixing techniques have a significant contribution to the cement tensile strength except for technique 1 and 3.


### Exercises 3.08 and 3.09

Reconsider the experiment in Problem 3.07.

* Rework part (3) of Problem 3.07 using Tukey’s test	with 	$\alpha = 0.05$. Do you get the same conclusions from Tukey’s test that you did from the graphical procedure and/or the Fisher LSD method?
* Explain the difference between the Tukey and Fisher procedures.
*Find a 95percent confidence interval on the mean tensile strength of the Portland cement produced by each of the four mixing techniques. Also find a 95 percent confidence interval on the difference in means for techniques 1 and 3. Does this aid you in interpreting the results of the experiment?

```r
# Applying Tukey's HSD test
tukey_result <- glht(anova_result, linfct=mcp(tech="Tukey"))
summary(tukey_result)
```

**Results:**
According to Tukeys test, we reject the null hypothesis at the $α=0.05$ level for the pair 1-4, 2-4, 3-4. Fisher LSD test at the same significance level indicated significant differences between all mixing techniques except 1-3. The Tukeys test is consistent with the observation in the previous figure and there is a highly significant difference between the techniques labeled by the Tukeys test. However, it can be seen from the figure that there should also be a large difference between techniques 2-3 and 1-2, which are only marked by the Fisher LSD test. Thus, the Tukeys test does not provide identical results when compared to the Fisher LSD test.

**Difference between Tukey and Fisher procedures:**
The primary difference between the Tukey and Fisher procedures lies in the control of the family-wise error rate (=the probability of making one or more Type I errors when performing multiple hypotheses tests):

*Fisher's LSD:* After finding a significant F-test in an ANOVA, it makes pairwise comparisons using the same error rate α. While it's more powerful, it doesn't adjust for multiple comparisons. This means it can inflate the familywise error rate, increasing the likelihood of Type I errors.

*Tukey's HSD:* This method specifically adjusts for multiple comparisons, controlling the familywise error rate at α. It's less powerful than LSD, meaning it might miss some true differences, but it's more conservative.

```r
CI_tukey_result <- confint(tukey_result)
print(CI_tukey_result)
```

```r
CI_tukey_df <- data.frame(
  label = c("2-1", "3-1", "4-1", "3-2", "4-2", "4-3"),
  estimate = c(CI_tukey_result$confint[,1]),
  lower = c(CI_tukey_result$confint[,2]),
  upper = c(CI_tukey_result$confint[,3])
)

ggplot(CI_tukey_df, aes(x=label, y=estimate)) +
  geom_errorbar(aes(ymin=lower, ymax=upper), width=0.2, color="red") +
  geom_point(size=3, color="blue") +
  coord_flip() +
  labs(x="Compared pair", y="Estimated Values", title="Estimated mean differences with Confidence Intervals") +
  theme_minimal()
```

It can be seen from the graph that only mixing techniques 1-3 are very close to zero difference in means. Furthermore, we can see that the confidence intervals for the difference in mean tensile strengths when using mixing techniques 3-2, 3-1, and 2-1 contain the value of zero.
Thus, we can say that for these observations, at the significance level $α=0.05$, it cannot be excluded that the difference in mean tensile strengths using these techniques is statistically insignificant. For other pairs the confidence interval the difference between two means does not contain 0, this means the difference between these two means is statistically significant at the 95% confidence level.

### Exercises 3.10

A product developer is investigating the tensile strength
of a new synthetic fiber that will be used to make cloth for
men’s shirts. Strength is usually affected by the percentage of
cotton used in the blend of materials for the fiber. The engineer
conducts a completely randomized experiment with five levels
of cotton content and replicates the experiment five times.


* Is there evidence to support the claim that cotton content
affects the mean tensile strength? Use $\alpha = 0.05$.
* Use the Fisher LSD method to make comparisons
between the pairs of means. What conclusions can you
draw?
* Analyze the residuals from this experiment and comment
on model adequacy.


```r
Ex03_10 <- read.table("https://raw.githubusercontent.com/francji1/01NAEX/main/data/Ex03_10.csv",header=TRUE,sep=";")
head(Ex03_10)
str(Ex03_10)
```

```r
data2 <- Ex03_10 %>% transmute(percentage = as.factor(Cotton_Weight),
                          strength = Observations)
```

```r
ggplot(data=data2, aes(x=percentage, y=strength)) +
  geom_boxplot() +
  stat_summary(fun.y="mean", size=1, color='purple') +
  geom_point(size=3, shape=16, color='red')
```

```r
# Conducting the ANOVA test
anova_result2 <- aov(strength ~ percentage, data=data2)
summary(anova_result2)
```

Based on the ANOVA test, the percentage of cotton contained in the material has a significant effect on the strength of the material.

```r
lsd_result2 <- LSD.test(anova_result2, "percentage", group=FALSE)
print(lsd_result2)
```

At the $α=0.05$ level, all groups except 15-35 and 20-25 have significantly different mean strengths.

```r
qqnorm(resid(anova_result2))
qqline(resid(anova_result2))
```

According to the QQ plot, the residuals do not appear to have an approximately normal distribution.

```r
plot(fitted(anova_result2), resid(anova_result2))
abline(h=0, col="red")
```

There are no clear patterns or funnel-like shapes in the dispersion of the residues, suggesting that the residues have a rather constant dispersion.

```r
plot(resid(anova_result2), type="p", main="Residuals in Order")
```

No clear patterns are seen in the graph, suggesting that the residuals are independent.

The assumption of a normal distribution of residuals does not seem to hold, so the model cannot be considered adequate for the data. The problem could be solved by transforming the data.
